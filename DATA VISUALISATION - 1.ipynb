```python
catg_counts = data['Genre'].value_counts()
catg_counts
```

```python
plt.plot(x,y)
plt.xticks(rotation = 90)
plt.show()
```

  # Univariate Analysis

```python
# we are doing analysis on just one variable Genres

x = catg_counts.index
y = catg_counts
plt.figure(figsize=(10,8))
plt.bar(x,y,width = 0.8,color = 'purple')
plt.xticks(rotation = 90)
plt.xlabel('Genres')
plt.ylabel('Counts')
plt.show()
```

```python
# Same Process in Seaborn
# we don't have to calculate any counts here, we don't have to take y axis.
# Seborn automatically does that for us, This is how advance seaborn is.
```

```python
sns.countplot(x= data['Genre'],color = 'Pink')
plt.xticks(rotation = 90)
plt.show()

# But this is not sorted?
```

```python
sns.countplot(data=data,x='Genre',
             order = data['Genre'].value_counts().index,
             color = 'blue')
plt.xticks(rotation = 90)
plt.show()

# Now it is sorted
```

```python
# Can we also create horizontal plot?

sns.countplot(data=data,y='Genre',
             color = 'Yellow')
```

## Univariate for Continuous/Numerical

```python
# It is giving me the frequency just like countplot.
# So why aren't we using countplot here?
# Because the variable here is Continuous.
# sns.countplot is good with categorical type of data

plt.hist(data['Year'])
plt.xlabel('Years')
plt.ylabel('Counts')
plt.xticks(rotation = 90)
plt.show()

# just like we used to create bins in pandas using pd.cut(It is creating bins automatically here)
# what if we want bins according to us?
```

```python
plt.hist(data['Year'],bins = 4)

# Here bins are nothing but a list of ndimensional array.
```

```python
# According to this plot, The maximum Games were released in between 2000-2010.
```

```python
# How do we do this using sns? And to make our plot more beautiful
```

```python
sns.histplot(data['Year'],bins = 4)

# This is more beautifully designed
# we don't have to add labels here. 
```
