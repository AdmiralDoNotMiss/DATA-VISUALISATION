# Box whiskers Plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('C:/Users/htc/final_vg.csv')
data
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(data = data, x = 'Global_Sales')
plt.show()
```

# Bivariate Analysis

```python
data['Name'].value_counts()
```

```python
ih = data.loc[data['Name'] == 'Ice Hockey']
ih
```

```python
# let's create a simple line chart first
```

```python
sns.lineplot(data = ih, x = 'Year', y = 'Global_Sales')
```

```python
# we can clearly see the trend for 1-2 years from 1985- 1987 there was a sudden boom in sales of icehockey game.
# After that there is very low sales constantly for years
```

```python
sns.lineplot(data = ih, x = 'Year', y = 'NA_Sales')
```

```python
# Let's take one more game Baseball
```

```python
bb = data.loc[data['Name'] == 'Baseball']
bb
```

```python
sns.lineplot(data = bb, x = 'Year', y = 'NA_Sales')
```

```python
# let's compare both of the games NA Sales in a single chart
```

```python
sns.lineplot(data = ih , x = 'Year', y ='NA_Sales', label = 'IceHockey')
sns.lineplot(data = bb , x = 'Year', y ='NA_Sales', label = 'Baseball')
plt.title('NA_Sales vs Years')
plt.show()

# This is a multi variate analysis, We are working with 3 features here Year, Sales, Name
```

```python
# Also we can put limit to our data. That will actually zoom in the chart for us.

sns.lineplot(data = ih , x = 'Year', y ='NA_Sales', label = 'IceHockey')
sns.lineplot(data = bb , x = 'Year', y ='NA_Sales', label = 'Baseball')
plt.title('NA_Sales vs Years')
plt.xlim(left = 1990, right = 2010)
plt.ylim(bottom = 0.6, top = 0.9)
plt.show()
```

```python
# Another way to put legends
```

```python
sns.lineplot(data = ih , x = 'Year', y ='NA_Sales')
sns.lineplot(data = bb , x = 'Year', y ='NA_Sales')
plt.title('NA_Sales vs Years')
plt.legend(['Icehockey','Baseball'])
plt.show()
```

```python
# we can also change the location of our legend
```

```python
sns.lineplot(data = ih , x = 'Year', y ='NA_Sales')
sns.lineplot(data = bb , x = 'Year', y ='NA_Sales')
plt.title('NA_Sales vs Years')
plt.legend(['Icehockey','Baseball'],loc = 'upper left')
plt.show()
```

```python
# Also we can put our legend outside the chart
```

```python
sns.lineplot(data = ih , x = 'Year', y ='NA_Sales')
sns.lineplot(data = bb , x = 'Year', y ='NA_Sales')
plt.title('NA_Sales vs Years')
plt.legend(['Icehockey','Baseball'],loc = (1,0.8))
plt.show()
```

```python
# let's fetch our top 3 publishers, top 3 platform, top 3 genre
```

```python
top_3_pub = data['Publisher'].value_counts().index[:3]
top_3_pub
```

```python
top_3_platform = data['Platform'].value_counts().index[:3]
top_3_platform
```

```python
top_3_genre = data['Genre'].value_counts().index[:3]
top_3_genre
```

```python
top_3_data = data.loc[(data['Publisher'].isin(top_3_pub)) & 
                     (data['Platform'].isin(top_3_platform))&
                     (data['Genre'].isin(top_3_genre))]
```

```python
top_3_data
```

```python
# Now we will work with these datapoints only
```

# Scatter plot

```python
sns.scatterplot(data = top_3_data, x = 'NA_Sales', y = 'Global_Sales')
```

```python
# Broadly 2 things we can understand from scatter plots.
# 1. Are the two variables correlated with each other, Basically what's the trend between them
# 2. Is there a positive trend or negative trend?
```

```python
# scatterplot in multivariate analysis
```

```python
sns.scatterplot(data = top_3_data, x = 'EU_Sales', y = 'Global_Sales', hue = 'Genre')

# This is a multivariate analysis, we are using genre in hue parameter 
```

```python
# Categorical - Categorical
```

```python
sns.countplot(data = top_3_data , x = 'Publisher', hue = 'Genre')
```

```python
# A DODGED BAR CHART IT IS
# LET'S DRAW SOME INFERENCES
# Namco Bandai Games mainly focuses on Action Genre
# Activision Publisher focuses on Action Genre as well
# However EA mostly focuses on Sports Genre
```

```python
sns.boxplot(data = top_3_data, x = 'Publisher', y = 'Global_Sales')
```

```python
# Let's draw some inferences from here
# Highest median Sales is of EA
# Highest spread is of Activision
# while namco bandai is least spreaded more of like concentrated sales
```

```python
sns.violinplot(data = top_3_data, x = 'Publisher', y = 'Global_Sales')

# Here violin plot is nothing but something like kdeplot + boxplot.
# It is showing us the distribution
```

```python
# Let's draw some inferences
# Namco bandai games is mostly distributed near its median Sales
# EA as well mostly distributed near its median Sales
# while Activision is distributed in two places 
```

```python
sns.barplot(data = top_3_data, 
           x = 'Platform',
           y = 'EU_Sales')

# by default it is giving us mean sales
```

```python
plt.figure(figsize = (15,10))
sns.boxplot(data = top_3_data, x = 'Publisher', y = 'Global_Sales', hue = 'Genre')
```

```python
# A question here. which games has been sold mostly by each company
# Namco bandai sold Most of the games in Action Genre
# Activision sold Most of the games in Misc Genre
# EA  sold Most of the games in Sports Genre
```

# Sub plots

```python
fig = plt.figure(figsize = (12,10))
plt.subplot(2,3,1)
sns.scatterplot(data= top_3_data, x = 'JP_Sales', y = 'Global_Sales', color = 'red')
plt.title('JP Sales vs Global Sales')
plt.subplot(2,3,3)
sns.scatterplot(data= top_3_data, x = 'EU_Sales', y = 'Global_Sales', color = 'pink')
plt.title('EU Sales vs Global Sales')
plt.subplot(2,3,5)
sns.scatterplot(data= top_3_data, x = 'NA_Sales', y = 'Global_Sales', color = 'blue')
plt.title('NA Sales vs Global Sales')
plt.show()
```
